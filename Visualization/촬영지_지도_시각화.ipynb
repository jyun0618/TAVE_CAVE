```python
import re
# 2. 위경도 문자열이 들어간 컬럼 찾기 (예: DESCRIPTION 열 또는 다른 곳)
# 여러 열에 들어갈 수 있으므로 모든 열을 검사해보자
def extract_coordinates(row):
    for col in row.index:
        val = str(row[col])
        if 'N' in val and 'E' in val:
            match = re.search(r'N([0-9.\-]+),\s*E([0-9.\-]+)', val)
            if match:
                return float(match.group(1)), float(match.group(2))
    return None, None

# 3. 위경도 추출 적용
df1['latitude'], df1['longitude'] = zip(*df1.apply(extract_coordinates, axis=1))

# 4. place_name 정리 (df1의 PLACENAME을 기준으로)
df1 = df1.rename(columns={'PLACENAME': 'place_name'})
df1 = df1[['place_name', 'latitude', 'longitude']].dropna()

# 5. 위경도 데이터프레임 만들기 (df1 + df2)
df2 = df2.rename(columns={'장소명': 'place_name', '위도': 'latitude', '경도': 'longitude'})
df2 = df2[['place_name', 'latitude', 'longitude']]

# 6. 병합할 모든 위경도 모으기
geo_df = pd.concat([df1, df2], ignore_index=True)
geo_df = geo_df.drop_duplicates(subset='place_name')

# 7. 리뷰 데이터와 병합
merged_df = pd.merge(df, geo_df, how='left', on='place_name')
```

```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd

# 데이터 불러오기
df = pd.read_csv('/content/drive/MyDrive/문화data/drama_locations_for_map.csv')

# 지도 중심 좌표 (대한민국 중심 정도)
map_center = [36.5, 127.8]
m = folium.Map(location=map_center, zoom_start=7, tiles='CartoDB positron')

# 마커 클러스터 설정
marker_cluster = MarkerCluster().add_to(m)

# 드라마별 다른 색상 지정
from itertools import cycle
import random

dramas = df['mediatitle'].unique()
colors = cycle([
    'red', 'blue', 'green', 'purple', 'orange', 'darkred', 'lightred',
    'beige', 'darkblue', 'darkgreen', 'cadetblue', 'darkpurple', 'white',
    'pink', 'lightblue', 'lightgreen', 'gray', 'black', 'lightgray'
])
color_map = {drama: next(colors) for drama in dramas}

# 지도에 마커 추가
for _, row in df.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"<b>{row['mediatitle']}</b><br>{row['place_name']}<br>{row['address']}",
        icon=folium.Icon(color=color_map[row['mediatitle']], icon='film')
    ).add_to(marker_cluster)

# 결과 저장
m.save('/content/drive/MyDrive/문화data/drama_locations_map.html')
```

드라마별 선택 가능하게 수정 코드

```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd

# CSV 불러오기
df = pd.read_csv('/content/drive/MyDrive/문화data/drama_locations_for_map.csv')

# 지도 초기화
m = folium.Map(location=[36.5, 127.8], zoom_start=7, tiles='CartoDB positron')

# 드라마별 색상 할당
from itertools import cycle
colors = cycle([
    'red', 'blue', 'green', 'purple', 'orange', 'darkred', 'lightred',
    'beige', 'darkblue', 'darkgreen', 'cadetblue', 'darkpurple', 'white',
    'pink', 'lightblue', 'lightgreen', 'gray', 'black', 'lightgray'
])
color_map = {}

# 드라마별 FeatureGroup 생성
for drama in df['mediatitle'].unique():
    group = folium.FeatureGroup(name=drama)
    color = next(colors)
    color_map[drama] = color

    for _, row in df[df['mediatitle'] == drama].iterrows():
        folium.Marker(
            location=[row['latitude'], row['longitude']],
            popup=folium.Popup(f"<b>{row['mediatitle']}</b><br>{row['place_name']}<br>{row['address']}", max_width=300),
            icon=folium.Icon(color=color, icon='film', prefix='fa')
        ).add_to(group)

    group.add_to(m)

# Layer Control 추가
folium.LayerControl(collapsed=False).add_to(m)

# 저장
output_path = "/content/drive/MyDrive/문화data/drama_layered_map.html"
m.save(output_path)
```
